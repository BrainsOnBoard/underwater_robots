```python
# Plot x diffs across different image sets

import numpy as np

dirnames = ['images1', 'images2', 'full bath_20cm']

diffs = []
for dirname in dirnames:
    images = load_images(dirname)
    diffs.append(get_diffs(dirname, images)[0])

for d in diffs:
    ys = np.linspace(0, 1, len(d))
    if d[-1] < 0:
        d = [-val for val in d]
    plt.plot(ys, d)
plt.show()
```
